Given the network structure where:

- $a_i = \sum_j w_j^i z_j$ for neurons 2, 3, 4, and 1
- $z_i = \text{sigmoid}(a_i)$ for $i = 2, 3, 4$

**Forward Pass Equations:**

$$z_5 = x \quad \text{(input)}$$

**Hidden Layer:**

$$a_2 = w_5^2 \cdot z_5$$

$$z_2 = \text{sigmoid}(a_2) = \frac{1}{1 + e^{-a_2}}$$

$$a_3 = w_5^3 \cdot z_5$$

$$z_3 = \text{sigmoid}(a_3) = \frac{1}{1 + e^{-a_3}}$$

$$a_4 = w_5^4 \cdot z_5$$

$$z_4 = \text{sigmoid}(a_4) = \frac{1}{1 + e^{-a_4}}$$

**Output Layer:**

$$a_1 = w_2^1 \cdot z_2 + w_3^1 \cdot z_3 + w_4^1 \cdot z_4$$

$$z_1 = \text{sigmoid}(a_1) = \frac{1}{1 + e^{-a_1}}$$

**Complete Function:**

$$f(x) = \text{sigmoid}(w_2^1 \cdot \text{sigmoid}(w_5^2 \cdot x) + w_3^1 \cdot \text{sigmoid}(w_5^3 \cdot x) + w_4^1 \cdot \text{sigmoid}(w_5^4 \cdot x))$$


### Part b

(b) Deduce the equation to calculate $_i$ (the error value per neuron) for all the neurons.
Write a function that given a training sample and the weights of the network calculate $_i$ for each neuron.

**Hint:**

#### SIGMOID ####
$$f(x) = \dfrac{1}{1+e^{-x}} $$
$$df(x)/dx = f(x)(1-f(x))$$

**Answer:**
(put your math as embedded latex, or otherwise neatly formatted here)

Using the chain rule for backpropagation, we calculate $\delta_i = \frac{\partial E}{\partial a_i}$ for each neuron.

**Output Layer (Neuron 1):**

$$\delta_1 = \frac{\partial E}{\partial a_1} = \frac{\partial E}{\partial z_1} \cdot \frac{\partial z_1}{\partial a_1}$$

$$\delta_1 = -(target - z_1) \cdot z_1(1 - z_1)$$

**Hidden Layer (Neurons 2, 3, 4):**

For neuron 2:
$$\delta_2 = \frac{\partial E}{\partial a_2} = \frac{\partial E}{\partial z_2} \cdot \frac{\partial z_2}{\partial a_2} = \frac{\partial E}{\partial a_1} \cdot \frac{\partial a_1}{\partial z_2} \cdot \frac{\partial z_2}{\partial a_2}$$

$$\delta_2 = \delta_1 \cdot w_2^1 \cdot z_2(1 - z_2)$$

For neuron 3:
$$\delta_3 = \delta_1 \cdot w_3^1 \cdot z_3(1 - z_3)$$

For neuron 4:
$$\delta_4 = \delta_1 \cdot w_4^1 \cdot z_4(1 - z_4)$$

**Summary:**
- $\delta_1 = -(target - z_1) \cdot z_1(1 - z_1)$
- $\delta_2 = \delta_1 \cdot w_2^1 \cdot z_2(1 - z_2)$
- $\delta_3 = \delta_1 \cdot w_3^1 \cdot z_3(1 - z_3)$
- $\delta_4 = \delta_1 \cdot w_4^1 \cdot z_4(1 - z_4)$
- $\delta_5 = 0$ (input neuron has no error)

Plot the evolution of the error and the final predictions of the trained network. Write down the weights of the trained nettwork.

```python
##**Answer:**

import numpy as np 

# defining the backprop for Sigmoid Function
def back_sigmoid(x):
  return x*(1-x)

# defining the Sigmoid Function
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def forward_pass(X, weights_input_hidden, weights_hidden_output):
    # Compute hidden layer activations
    # X is of shape (N, 1), weights_input_hidden is (1, 3)
    # hidden_weighted_sum will be (N, 3)
    hidden_weighted_sum = X @ weights_input_hidden
    hiddenLayer_activations = sigmoid(hidden_weighted_sum)
    
    # Compute output layer
    # hiddenLayer_activations is (N, 3), weights_hidden_output is (3, 1)
    # output_weighted_sum will be (N, 1)
    output_weighted_sum = hiddenLayer_activations @ weights_hidden_output
    output = sigmoid(output_weighted_sum)
    
    ## output is of shape (N, 1)
    ## hiddenLayer_activations is of shape (N, 3)
    return output, hiddenLayer_activations

def backward_pass(X, y, output, weights_hidden_output, weights_input_hidden, hiddenLayer_activations):
  # calculating rate of change of error w.r.t weight between hidden and output layer
  
  # Compute delta for output layer (neuron 1)
  # delta_1 = -(target - z_1) * z_1 * (1 - z_1)
  delta_output = -(y - output) * back_sigmoid(output)  # Shape: (N, 1)
  
  # Compute gradient for weights_hidden_output
  # gradient = hiddenLayer_activations^T @ delta_output, then average over all samples
  error_wrt_weights_hidden_output = (hiddenLayer_activations.T @ delta_output) / X.shape[0]  # Shape: (3, 1)
  
  # Compute delta for hidden layer (neurons 2, 3, 4)
  # delta_hidden = delta_output @ weights_hidden_output^T * z_hidden * (1 - z_hidden)
  delta_hidden = (delta_output @ weights_hidden_output.T) * back_sigmoid(hiddenLayer_activations)  # Shape: (N, 3)
  
  # Compute gradient for weights_input_hidden
  # gradient = X^T @ delta_hidden, then average over all samples
  error_wrt_weights_input_hidden = (X.T @ delta_hidden) / X.shape[0]  # Shape: (1, 3)
    
  return error_wrt_weights_hidden_output, error_wrt_weights_input_hidden


def train(X_train, y_train):

    # defining the model architecture
    inputLayer_neurons = 1  # number of neurons at input
    hiddenLayer_neurons = 3  # number of hidden layers neurons
    outputLayer_neurons = 1  # number of neurons at output layer

    # initializing weight
    weights_input_hidden = np.random.uniform(size=(inputLayer_neurons, hiddenLayer_neurons))
    weights_hidden_output = np.random.uniform(
        size=(hiddenLayer_neurons, outputLayer_neurons)
    )
    # defining the parameters
    lr = 0.1
    epochs = 1000

    losses =  []

    for ep in range(epochs):
      output_, hiddenLayer_activations = forward_pass(X_train, weights_input_hidden, weights_hidden_output)

      ## Backward Propagation
      # calculating error
      error = np.square(y_train - output_) / 2

      error_wrt_weights_hidden_output, error_wrt_weights_input_hidden = backward_pass(X_train, y_train, 
                                                                                      output_, 
                                                                                      weights_hidden_output, 
                                                                                      weights_input_hidden, 
                                                                                      hiddenLayer_activations)

      # updating the weights
      weights_hidden_output = weights_hidden_output - lr * error_wrt_weights_hidden_output
      weights_input_hidden = weights_input_hidden - lr * error_wrt_weights_input_hidden

      # print error at every 100th epoch
      epoch_loss = np.average(error)
      if ep % 100 == 0:
          print(f"Error at epoch {ep} is {epoch_loss:.5f}")

      # appending the error of each epoch
      losses.append(epoch_loss)

    import matplotlib.pyplot as plt
    plt.plot(losses)
    plt.show()

    final_pred, _ = forward_pass(X_train, weights_input_hidden, weights_hidden_output)

    plt.plot(final_pred, c='r')
    plt.plot(y_train, c='b')
    plt.show()

    print("Weights of the network are: ")
    print("w13, w12, w14", weights_hidden_output.T)
    print("w35, w25, w45", weights_input_hidden)


# defining training data
X_train = np.zeros((11, 1)).astype(np.float32)
y_train = np.zeros((11, 1)).astype(np.float32)

X_train[0] = -3
X_train[1] = -2
X_train[2] = -1.5
X_train[3] = -1.0
X_train[4] = -0.5
X_train[5] = 0.0
X_train[6] = 0.5
X_train[7] = 1.0
X_train[8] = 1.5
X_train[9] = 2.0
X_train[10] = 3.0


y_train[0] = 0.7312
y_train[1] = 0.7339
y_train[2] = 0.7438
y_train[3] = 0.7832
y_train[4] = 0.8903
y_train[5] = 0.9820
y_train[6] = 0.8114
y_train[7] = 0.5937
y_train[8] = 0.5219
y_train[9] = 0.5049
y_train[10] = 0.5002


train(X_train, y_train)
```
